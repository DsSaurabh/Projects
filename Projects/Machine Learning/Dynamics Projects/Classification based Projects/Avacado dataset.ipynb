# 1-Regression

```python
df1.head()
```

### Converting categorical variables into numeric types.

```python
#We will convert all categorical variables into numeric types using Label Encoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

#New Variable for Region
df1['Region']=le.fit_transform(df1['region'])

col=['type','Region']
for i in col:
    df1[i]=le.fit_transform(df1[i])
```

```python
df1.head()
```

```python
#Now create a dummy variable

df1=pd.get_dummies(df1,columns=['type','Region'])
df1.head()
```

## Dropping Columns

```python
model=df1.copy()
```

```python
model.drop ( ["Date",'region'] , inplace = True , axis = 1)
model.head()
```

Splitting the data into input and target variable.

```python
x=model.drop(['AveragePrice'],axis=1)
x.shape
```

```python
y=model['AveragePrice']
y.shape
```

```python
#Scaling the input variable

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
#Breaking our input and target variable into train and test data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
#Checking the shape
print(x_train.shape,'\t',x_test.shape)
```

```python
#Checking the shape
print(y_train.shape,'\t',y_test.shape)
```

```python
#Importing our Models Library

from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor

#Importing Error Metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
#Using algorithms via for loop

models=[LinearRegression(),DecisionTreeRegressor(),KNeighborsRegressor(),SVR(),Lasso(),Ridge(),ElasticNet()]
for m in models:
    m.fit(x_train,y_train)
    print('Score of',m,'is:',m.score(x_train,y_train))
    predm=m.predict(x_test)
    print('Error:')
    print('Mean absolute error:',mean_absolute_error(y_test,predm))
    print('Mean squared error:',mean_squared_error(y_test,predm))
    print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predm)))
    print("r2_score:",r2_score(y_test,predm))
    print('************************************************************************************************')
    print('\n')
```

```python
#Cross Validate the models

from sklearn.model_selection import cross_val_score
models=[LinearRegression(),DecisionTreeRegressor(),KNeighborsRegressor(),SVR(),Lasso(),Ridge(),ElasticNet()]
for m in models:
    score=cross_val_score(m,x,y,cv=15,scoring='r2')
    print('Score of',m,'is:')
    print("Score:",score)
    print('Mean Score:',score.mean())
    print("Standard deviation:",score.std())
    print('************************************************************************************************')
    print('\n')
```

Kneighbors Regressor gives best score.

```python
#Using KNeighbors model because it gives the best result

knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
print('Score:',knn.score(x_train,y_train))
predknn=knn.predict(x_test)
print('\n')
print('Mean absolute error:',mean_absolute_error(y_test,predknn))
print('Mean squared error:',mean_squared_error(y_test,predknn))
print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predknn)))
print('\n')
print("r2_score:",r2_score(y_test,predknn))
```

```python
#Cross Validate KNeighbors

score=cross_val_score(knn,x,y,cv=5,scoring='r2')
print("Score:",score)
print('Mean Score:',score.mean())
print("Standard deviation:",score.std())
```

# Using Ensemble Technique to Boostup our score

```python
#Using Random Forest

from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=40,random_state=42)
rf.fit(x_train,y_train)
print('Score:',rf.score(x_train,y_train))
predrf=rf.predict(x_test)
print('\n')
print('Mean absolute error:',mean_absolute_error(y_test,predrf))
print('Mean squared error:',mean_squared_error(y_test,predrf))
print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predrf)))
print('\n')
print("r2_score:",r2_score(y_test,predrf))
```

```python
#Using Ada Boost

from sklearn.ensemble import AdaBoostRegressor
lasso=Lasso()

adb=AdaBoostRegressor(base_estimator=knn,n_estimators=20,learning_rate=0.1,random_state=59)
adb.fit(x_train,y_train)
print('Score:',adb.score(x_train,y_train))
predadb=adb.predict(x_test)
print('\n')
print('Mean absolute error:',mean_absolute_error(y_test,predadb))
print('Mean squared error:',mean_squared_error(y_test,predadb))
print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predadb)))
print('\n')
print("r2_score:",r2_score(y_test,predadb))
```

### Random Forest Regressor gives best score among all the models.

```python
#Plotting scatter plot between test data and predicted data for Random Forest

plt.scatter(y_test,predrf)
plt.xlabel('y_test')
plt.ylabel('Predicted Y')
plt.title("Scatter plot between test data and predicted data")
plt.show()
```

```python
#Cross Validate Random Forest Regressor

score=cross_val_score(rf,x,y,cv=5,scoring='r2')
print("Score:",score)
print('Mean Score:',score.mean())
print("Standard deviation:",score.std())
```

Random Forest is working well because it gives the best r2_score and also it is not suffering from overfitting and

# 2-Classification

```python
df2.head()
```

```python
#Making Copy of our dataset
dataset=df2.copy()
```

Splitting the data into input and target variable.

```python
x=dataset.drop(['Date','region','type'],axis=1)
x.shape
```

```python
y=df1['region']
y.shape
```

```python
#Our target variable is in object form so change it in numeric form.

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
y
```

```python
y=y.reshape(-1,1)
```

```python
y.shape
```

```python
#Shape Checking
print(x.shape,'\t',y.shape)
```

```python
#Scaling the input variable

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
#Breaking our target and input variable into training and test data.

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.22,random_state=42)
```

```python
#Checking shape

print(x_train.shape,'\t',x_test.shape)
```

```python
#Checking shape

print(y_train.shape,'\t',y_test.shape)
```

```python
#Importing all the model Library

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

#Importing error metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# All Algorithm by using for loop

model=[LogisticRegression(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier()]


for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print('Accuracy score of',m,'is:')
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('\n')
```

```python
#Cross Validate the models

from sklearn.model_selection import cross_val_score
models=[LogisticRegression(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier()]
for m in models:
    score=cross_val_score(m,x,y,cv=15,scoring='accuracy')
    print("Model:",m)
    print("Score:",score)
    print('Mean Score:',score.mean())
    print("Standard deviation:",score.std())
    print('************************************************************************************************')
    print('\n')
```

#### SVC and KNeighbors Classifier Gives best score. so we cross check boths models one by one to get the best model.

```python
# Using KNeighborsClassifier Model

knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
knn.score(x_train,y_train)
predknn=knn.predict(x_test)
print(accuracy_score(y_test,predknn))
print(confusion_matrix(y_test,predknn))
print(classification_report(y_test,predknn))
```

```python
#Using GridSearchCV to find out the best parameter in SVM

from sklearn.model_selection import GridSearchCV

parameters={'kernel':('linear','rbf','poly'),'C':[1,10],'random_state':range(42,100)}
sv=SVC()

clf=GridSearchCV(sv,parameters)
clf.fit(x,y)
print(clf.best_params_)
```

```python
# Using SVM with best result

sv=SVC(kernel='linear',C=10,random_state=42)
sv.fit(x_train,y_train)
sv.score(x_train,y_train)
predsv=sv.predict(x_test)
print(accuracy_score(y_test,predsv))
print(confusion_matrix(y_test,predsv))
print(classification_report(y_test,predsv))
```

### Among both models SVC gives best score.

## Using Ensemble Technique to Boostup our score.

```python
#Using Random Forest
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier(n_estimators=120,random_state=42)
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
print(accuracy_score(y_test,predrf))
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))
```

```python
#Using Ada Boost
from sklearn.ensemble import AdaBoostClassifier

ad=AdaBoostClassifier(base_estimator=sv,n_estimators=60,random_state=42,algorithm='SAMME')
ad.fit(x_train,y_train)
predad=ad.predict(x_test)
print(accuracy_score(y_test,predad))
print(confusion_matrix(y_test,predad))
print(classification_report(y_test,predad))
```

### Among all Random Forest Classifier is giving best result then SVC is giving best result.
