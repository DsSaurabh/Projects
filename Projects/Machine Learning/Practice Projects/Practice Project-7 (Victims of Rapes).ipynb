```python
#Copying the dataset
df=rape_victims.copy()
```

```python
#Dropping the Area_Name and Subgroup because it is not helpful in model training

df.drop(columns=['Area_Name','Subgroup'],axis=1,inplace=True)
df.columns
```

# Removing Outliers

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
df_new=df[(z<3).all(axis=1)]
```

```python
print(df.shape,'\t',df_new.shape)
```

```python
df=df_new
df.shape
```

# Model Training

```python
df.columns
```

```python
df_x=df.drop('Victims_of_Rape_Total',axis=1)
df_x.shape
```

```python
y=df['Victims_of_Rape_Total']
y.shape
```

```python
#Scaling the input variable

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(df_x)
```

```python
x=pd.DataFrame(x,columns=df_x.columns)
x
```

```python
#Importing Library for splitting data into Training and Testing

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=.22)
```

```python
print(x_train.shape,'\t',x_test.shape)
```

```python
print(y_train.shape,'\t',y_test.shape)
```

```python
#Importing Models
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor

#Importing Metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
models=[LinearRegression(),Lasso(),Ridge(),ElasticNet(),SVR(),KNeighborsRegressor(),DecisionTreeRegressor()]

for m in models:
    m.fit(x_train,y_train)
    print('Score of',m,'is:',m.score(x_train,y_train))
    predm=m.predict(x_test)
    print('Error of',m,'is:')
    print('Mean absolute error:',mean_absolute_error(y_test,predm))
    print('Mean squared error:',mean_squared_error(y_test,predm))
    print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predm)))
    print("r2_score:",r2_score(y_test,predm))
    print('**************************************************************************************************')
    print('\n')
```

```python
#Linear Regression model is giving the best result

lm=LinearRegression()
lm.fit(x_train,y_train)
print("Score:",lm.score(x_train,y_train))
predlm=lm.predict(x_test)
print('\n')
print('Errors:')
print('Mean absolute error:',mean_absolute_error(y_test,predlm))
print('Mean squared error:',mean_squared_error(y_test,predlm))
print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predlm)))
print('\n')
print("r2_score:",r2_score(y_test,predlm))
```

```python
#Cross validate the model to avoid overfitting and under fitting
from sklearn.model_selection import cross_val_score

score=cross_val_score(lm,x,y,cv=5,scoring='r2')
print(score)
print(score.mean())
print(score.std())
```

```python
#Saving the model
from sklearn.externals import joblib

#Saving the file as pickle
joblib.dump(lm,'Rape-Victims.pkl')
```
