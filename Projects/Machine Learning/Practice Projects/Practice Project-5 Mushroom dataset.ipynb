```python
#The data is categorial so I convert it with LabelEncoder

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in data.columns:
    data[col] = le.fit_transform(data[col])
 
data.head()
```

```python
data.describe()
```

```python
#From the table above it can be seen that the column "veil-type" is 0 and not contributing to the data so I remove it.

data=data.drop(["veil-type"],axis=1)
print(data.shape)
data.head()
```

```python
#Correlation between the variables

plt.figure(figsize=(16,14))
sns.heatmap(data.corr(),cmap="YlGnBu", annot=True)
plt.show()
```

Seprating Input and target variable

```python
df_x=data.iloc[:,1:22]
df_x.head()
```

```python
df_y=data.iloc[:,0]
df_y.head()
```

```python
x=pd.DataFrame(data=df_x)
x
```

```python
y=pd.DataFrame(data=df_y)
y
```

```python
#Splitting the Data into Training and Testing set

from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=47)
```

```python
print(x_train.shape,'\t',x_test.shape)
```

```python
print(y_train.shape,'\t',y_test.shape)
```

Importing all the models library

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_curve,auc
```

```python
#Using Logistic Regression

lg=LogisticRegression()
lg.fit(x_train,y_train)
y_pred=lg.predict(x_test)
print('Accuracy Score:',accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
#Printing Confusion Matrix 

cfm=confusion_matrix(y_test,y_pred)

sns.heatmap(cfm,annot=True,cmap='Blues')
plt.title('Logistic Regression confusion matrix')
plt.xlabel('Predicted Value')
plt.ylabel('True Value')
```

```python
# SVM by using function

def svmkernel(ker):
    svc=SVC(kernel=ker)
    svc.fit(x_train,y_train)
    svc.score(x_train,y_train)
    predsvc=svc.predict(x_test)
    print(accuracy_score(y_test,predsvc))
    print(confusion_matrix(y_test,predsvc))
    print(classification_report(y_test,predsvc))
    
    
svmkernel('rbf')
print('\n')
svmkernel('linear')
print('\n')
svmkernel('poly')
```

```python
# Kernel=poly is best working
svc=SVC(kernel='poly')    
svc.fit(x_train,y_train)
svc.score(x_train,y_train)
predsvc=svc.predict(x_test)
print(accuracy_score(y_test,predsvc))
print(confusion_matrix(y_test,predsvc))
print(classification_report(y_test,predsvc))

#Printing Confusion Matrix 

cfm=confusion_matrix(y_test,predsvc)

sns.heatmap(cfm,annot=True,cmap='tab20')
plt.title('SVC for kernel poly confusion matrix')
plt.xlabel('Predicted Value')
plt.ylabel('True Value')
```

```python
# All Algorithm by using for loop

model=[LogisticRegression(),GaussianNB(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier()]


for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print('Accuracy score of',m,'is:')
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('**************************************************************************************************')
    print('\n')
```

```python
# Cross Validated all algorithms by using for loop

model=[GaussianNB(),LogisticRegression(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier()]


for m in model:
    score=cross_val_score(m,x,y,cv=10)
    print('Score of',m,'is:')
    print(score)
    print('Mean:',score.mean())
    print('Std:',score.std())
    print('*************************************************************************************************')
    print('\n')
```

```python
# Above all Decision Tree Classifier is working well
#DecisionTreeClassifier(criterion='gini')------------->default
#Two criterion ginni and entropy

dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
dtc.score(x_train,y_train)
preddtc=dtc.predict(x_test)
print(accuracy_score(y_test,preddtc))
print(confusion_matrix(y_test,preddtc))
print(classification_report(y_test,preddtc))
```

```python
#Printing Confusion Matrix 

cfm=confusion_matrix(y_test,preddtc)

sns.heatmap(cfm,annot=True,cmap='tab10')
plt.title('Decision Tree Classifier confusion matrix')
plt.xlabel('Predicted Value')
plt.ylabel('True Value')
```

```python
#Printing roc curve

from sklearn.metrics import roc_curve,roc_auc_score

y_pred_prob=dtc.predict_proba(x_test)[:,0]
fpr,tpr,thresholds=roc_curve(y_test,y_pred_prob)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='Decision tree Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("Decision Tree Classifier")
plt.show()
```

```python
#Printing auc_score
auc_score=roc_auc_score(y_test,dtc.predict(x_test))
auc_score
```

```python
#Using Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier(criterion='entropy', random_state=45,n_estimators=100)
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
print(accuracy_score(y_test,predrf))
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))
```

```python
#Printing Confusion Matrix 

cfm=confusion_matrix(y_test,predrf)

sns.heatmap(cfm,annot=True,cmap='tab10_r')
plt.title('Decision Tree Classifier confusion matrix')
plt.xlabel('Predicted Value')
plt.ylabel('True Value')
```

```python
#Printing roc curve

from sklearn.metrics import roc_curve,roc_auc_score

y_pred_prob=rf.predict_proba(x_test)[:,0]
fpr,tpr,thresholds=roc_curve(y_test,y_pred_prob)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='Decision tree Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("Random Forest Classifier")
plt.show()
```

```python
#Printing auc_score
auc_score=roc_auc_score(y_test,rf.predict(x_test))
auc_score
```

```python
#Here Random Forest and Decision Tree are working well.
#So cross validate both models to see overfitting

from sklearn.model_selection import cross_val_score
```

```python
#Cross validate Decision Tree

score=cross_val_score(dtc,x,y,cv=10,scoring='accuracy')
print(score)
print('Mean:',score.mean())
print('Std:',score.std())
```

```python
#Cross validate RandomForest

score=cross_val_score(rf,x,y,cv=10,scoring='accuracy')
print(score)
print('Mean:',score.mean())
print('Std:',score.std())
```
