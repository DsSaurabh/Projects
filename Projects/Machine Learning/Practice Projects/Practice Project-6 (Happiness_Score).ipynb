# Model Training

```python
df_x=score.drop(['Country','Region','Happiness Score','Happiness Rank'],axis=1)
df_x
```

```python
y=score['Happiness Score']
y
```

```python
#Scaling the data
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
x
```

```python
#Dividing the data into training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
print(x_train.shape,'\t',x_test.shape)
```

```python
print(y_train.shape,'\t',y_test.shape)
```

```python
#Importing the models library

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

#Importing the Error Metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
#Applying all models 

models=[LinearRegression(),DecisionTreeRegressor(),KNeighborsRegressor(),SVR(kernel='linear')]

for m in models:
    m.fit(x_train,y_train)
    score=m.score(x_train,y_train)
    print('Score of',m,'is:',score)
    y_pred=m.predict(x_test)
    print('Mean Absolute Error:',mean_absolute_error(y_test,y_pred))
    print('Mean Squared Error:',mean_squared_error(y_test,y_pred))
    print('RMSE:',np.sqrt(mean_squared_error(y_test,y_pred)))
    print('R2 score:',r2_score(y_test,y_pred))
    print('*****************************************************************************************')
    print('\n')
```

```python
#Linear Regression Model is working well
lm=LinearRegression()
lm.fit(x_train,y_train)
print("Score:",lm.score(x_train,y_train))
predlm=lm.predict(x_test)
print('\n')
print('Errors:')
print('Mean absolute error:',mean_absolute_error(y_test,predlm))
print('Mean squared error:',mean_squared_error(y_test,predlm))
print('Root Mean Squared Error:',np.sqrt(mean_squared_error(y_test,predlm)))
print('\n')
print("r2_score:",r2_score(y_test,predlm))
```

```python
#Cross validate the model to avoid overfitting and under fitting
from sklearn.model_selection import cross_val_score

score=cross_val_score(lm,x,y,cv=5,scoring='r2')
print(score)
print(score.mean())
print(score.std())
```

```python
#Make prediction
predictions=lm.predict(x_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('y_test')
plt.ylabel('Predicted Y')
```

The above graph shows that test value and predicted value are inline with each other. This means that our model
is good.
